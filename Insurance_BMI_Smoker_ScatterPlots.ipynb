```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# Path of the file to read
insurance_filepath = "F://datasets_116573_334386_insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data.head()
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python

sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'], ci = None, color="g", x_bins=100)
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data, ci = None, x_bins=100)
```

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```
